```python
class TrafficDevice:

    def activate(self):
        print("Activating general traffic device...")


class TrafficLight(TrafficDevice):

    def activate(self):
        print("Traffic Light: Controlling traffic flow (Red / Yellow / Green).")


class SpeedCamera(TrafficDevice):

    def activate(self):
        print("Speed Camera: Monitoring vehicle speeds and capturing infractions.")


class PedestrianSignal(TrafficDevice):

    def activate(self):
        print("Pedestrian Signal: Displaying WALK / DON'T WALK signs.")


class EmergencySiren(TrafficDevice):

    def activate(self):
        print(
            "Emergency Siren: Blaring high-decibel alert for emergency vehicles!"
        )


if __name__ == "__main__":
    # Instantiate one object of each class
    devices = [
        TrafficLight(),
        SpeedCamera(),
        PedestrianSignal(),
        EmergencySiren(),
    ]

    print("--- Activating Smart City Traffic System ---")
    for device in devices:
        device.activate()
```
